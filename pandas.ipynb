only used numerical  column

```python
import pandas as pd

import numpy as np

 

# Sample Data

data = {'Age': [19,20,21,18,14,23,24,100,95]}  # 120 and 150 are potential outliers

df = pd.DataFrame(data)

 

# Summary statistics

print(df.describe())

```

```python
Q1= df.quantile(0.25)
```

```python
Q3= df.quantile(0.75)
```

```python
IQR = Q3-Q1
```

```python
IQR
```

```python
LW = Q1 - 1.5 * IQR
print("lower wishker is :{LW})
      
```

```python
df
```

```python
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
```

```python
print("lower wishker is :" , lower_bound)

print("upper wishker is :" , upper_bound)
```

```python
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]


print(outliers)
```

```python
Q1 = df['Age'].quantile(0.25)

Q3 = df['Age'].quantile(0.75)

IQR = Q3 - Q1

 

lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR

 

print("Lower Whisker:", lower_bound)

print("Upper Whisker:" ,upper_bound)

 

outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

print(outliers)
```
