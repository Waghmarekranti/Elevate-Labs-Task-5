# Data Cleaning

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Name'] = df['Name'].str.strip()   
df['Name'] = df['Name'].str.title()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median())  
df['Fare'] = df['Fare'].astype(float)
```

```python
df['Ticket'] = df['Ticket'].astype(str).str.upper().str.strip()
```

```python
df['Name'] = df['Name'].str.replace(',', '', regex=True)
```

```python
df['Sex'] = df['Sex'].str.strip().str.capitalize()
```

```python
df.head()
```

# Data Preprocessing and modifications

```python
# Age group: Child, Young, Adult, Senior
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], 
                         labels=['Child','Teen','Young','Adult','Senior'])
```

```python
df['Status'] = 'Active'  # all rows will have 'Active'
```

```python
df.head()
```

# Visulation

## 1: Pairplot

```python
sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch', 'Pclass', 'Status']], 
             hue='Status')
plt.show()
```

Observations

Survivors usually paid higher fare.

Younger passengers had higher survival.

Pclass 1 (upper class) shows higher survival.

## 2 : Heatmap (Correlation Matrix)

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()

```

Observations

Fare and Pclass strongly negatively correlated (rich passengers = lower Pclass number).

Survival (Status) correlates with Fare and Pclass.

Age has weak correlation with survival.

## 3 : Histograms (Distribution Plots)

#### 3.1 : Age Distribution

```python
plt.hist(df['Age'], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Observation: Most passengers are between 2040 years.

#### 3.1 : Fare Distribution

```python
plt.hist(df['Fare'], bins=30)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

Observation: Fare is highly right-skewed (many low fares, few very high).

## 4 :  Boxplots

#### 4.1 :Age vs Sex

```python
sns.boxplot(x='Sex', y='Age', data=df)
plt.show()
```

Observation:

1st  class fares are much higher.

3rd  class has lowest fare range.

Observation: Male age distribution is slightly wider; no major outliers.

#### 4.2 : Fare vs Pclass

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.show()
```

Observation:

1st class fares are much higher.

3rd class has lowest fare range.

## 5 Scatterplots

#### 5.1 : Fare vs Age

```python
sns.scatterplot(x='Age', y='Fare', hue='Status', data=df)
plt.show()
```

Observation: Higher fare passengers were more likely to survive at all ages.

#### 5.2 : Pclass vs Age

```python
sns.scatterplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.show()
```

Observation: Class distribution is spread across age groups.

## 6 : Countplots (Categorical Visuals)

#### 6.1 : Survival Count

```python
sns.countplot(x='Status', data=df)
plt.show()
```

Observation: Non-survivors are more than survivors.

#### 6.2 : Embarked vs Status

```python
sns.countplot(x='Embarked', hue='Status', data=df)
plt.show()
```

Observation: Passengers from C (Cherbourg) survived more (wealthier passengers).

# Final Summary of Findings
###  1. Survival Factors

Higher Fare, 1st Class, and female passengers had significantly higher survival rates.

Children also show higher survival in pairplot.

###  2. Fare & Class

Strong relationship: 1st class paid much higher fares.

Fare distribution is right-skewed with many low-fare travelers.

###  3. Age Patterns

Majority passengers aged 2040.

Slight indication that younger people had better survival chances.

###  4. Correlation Heatmap

Fare  Pclass strongest correlation.

Survival correlates with Fare and Pclass but weakly with age.

###  5. Embarked

Most passengers embarked from Southampton, but survival was highest from Cherbourg.
